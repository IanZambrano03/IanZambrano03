```python
import pandas as pd
import numpy as np 
import matplotlib.pylab as plt 
import seaborn as sns
plt.style.use('ggplot')
```

# Con este grafico vamos a ver cuantos asambleistas de Construye sin han aportado

```python
data = [["Alexandra Campo Verde", 1], ["Ana Galarza", 1], ["Camilo Salinas", 8], ["En Grupo", 1], ["Maanuel Humberto Tapia", 1], ["Paul Fernando Carabajo", 4], ["Sandra Elizabeth Rueda Camacho", 2], ["Viviana Jacqueline Zambrano" , 1],["El Resto de Construye", 0]]
df = pd.DataFrame(data, columns=['Name', 'Value'])


names = df['Name']
values = df['Value']

# Step 4: Plot the data
plt.figure(figsize=(12, 6))
plt.bar(names, values, color='skyblue')
plt.xticks(rotation=45, ha='right')
plt.xlabel('Asambleistas')
plt.ylabel('Cantidad')
plt.title('Leyes promovidas por cada asambleistas desde Dic 2023 - Julio 2024')
plt.show()
```

# Ahora vamos a ver la cantidad de leyes que ha mandado Construye comparado con toda las leyes mandandos por los asambleistas

```python
import matplotlib.pyplot as plt


construye1 = ["Construye", 19]
asamblea = ["La Asamblea", 241]


labels = [construye1[0], asamblea[0]]
values = [construye1[1], asamblea[1]]


plt.figure(figsize=(8, 8))
plt.pie(values, labels=labels, autopct='%1.1f%%', startangle=140, colors=['lightblue', 'lightgreen'])
plt.title('% De Leyes Promovidas de Construye en comparacion con la Asamblea desde Dic 2023 - Julio 2024')
plt.axis('equal') 
plt.show()

```
